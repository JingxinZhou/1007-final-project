```python
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')
import numpy as np
from matplotlib import pyplot as plt
```

Read data of 4 brands

```python
raw_data_hd = pd.read_csv("haagendazs.csv")
raw_data_tt = pd.read_csv("talenti.csv")
raw_data_bj = pd.read_csv("benandjerrys.csv")
raw_data_by = pd.read_csv("breyers.csv")
```

Select relevant information - publish time and sentiment, and convert publish time to month

```python
data_hd = raw_data_hd[['published','sentiment']]
data_tt = raw_data_tt[['published','sentiment']]
data_bj = raw_data_bj[['published','sentiment']]
data_by = raw_data_by[['published','sentiment']]
```

```python
for i in range(len(raw_data_hd['published'])):
    timetmp = raw_data_hd['published'][i]
    p = time.strptime(timetmp, "%m/%d/%y %H:%M:%S %p") 
    data_hd['published'][i] = time.strftime("%Y-%m",p)
for i in range(len(raw_data_tt['published'])):
    timetmp = raw_data_tt['published'][i]
    p = time.strptime(timetmp, "%m/%d/%y %H:%M:%S %p") 
    data_tt['published'][i] = time.strftime("%Y-%m",p)
for i in range(len(raw_data_bj['published'])):
    timetmp = raw_data_bj['published'][i]
    p = time.strptime(timetmp, "%m/%d/%y %H:%M:%S %p") 
    data_bj['published'][i] = time.strftime("%Y-%m",p)
for i in range(len(raw_data_by['published'])):
    timetmp = raw_data_by['published'][i]
    p = time.strptime(timetmp, "%m/%d/%y %H:%M:%S %p") 
    data_by['published'][i] = time.strftime("%Y-%m",p)
```

```python
data_hd
```

```python
data_month_hd = data_hd.groupby('published').mean()
data_month_tt = data_tt.groupby('published').mean()
data_month_bj = data_bj.groupby('published').mean()
data_month_by = data_by.groupby('published').mean()
```

Plot the mean sentiment score of each month for each brand

```python
plt.figure(figsize=(15,5))
plt.plot(data_month_hd, label = 'haagendazs')
plt.plot(data_month_tt, label = 'talenti')
plt.plot(data_month_bj, label = 'benandjerrys')
plt.plot(data_month_by, label = 'breyers')
plt.legend()
plt.show()
```
