## Text Preprocessing
In this preprocessing function, I delete all stopwords symbols apart from letters, convert the text into lower case and stem or lemmatise the words.

Stopwords: A stop word is a commonly used word (such as “the”, “a”, “an”, “in”). We would not want these words to take up space in our dataset, or taking up valuable processing time. Also, wenn looking for common words in our dataset, finding out that "a" is the most common, is not particularly interesting

Lemmatising: Lemmatisation (or lemmatization) in linguistics is the process of grouping together the inflected forms of a word so they can be analysed as a single item, identified by the word's lemma, or dictionary form.\
For example: Loved -> love

```python
import re
from nltk.stem import WordNetLemmatizer, PorterStemmer, SnowballStemmer

stop_words_file = 'SmartStoplist.txt'

stop_words = []

with open(stop_words_file, "r") as f:
    for line in f:
        stop_words.extend(line.split()) 
        
stop_words = stop_words  

def preprocess(raw_text):
    
    #regular expression keeping only letters - more on them later
    letters_only_text = re.sub("[^a-zA-Z]", " ", raw_text)

    # convert to lower case and split into words -> convert string into list ( 'hello world' -> ['hello', 'world'])
    words = letters_only_text.lower().split()

    cleaned_words = []
    
    # remove stopwords
    for word in words:
        if word not in stop_words:
            cleaned_words.append(word)
    
    # lemmatise words
    lemmatized_words = []
    for word in cleaned_words:
        word = WordNetLemmatizer().lemmatize(word)
        lemmatized_words.append(word)
    
    # converting list back to string
    return " ".join(lemmatized_words)
```

```python
test_sentence = "this is a sentence to demonstrate how the preprocessing function works...!"

preprocess(test_sentence)
```

You can see that "works" was lemmatized to "work", all stop words and punctuation were removed. \
Let's apply that function to the incident texts in our ice cream dataframe

```python
dat_hd['content'] = dat_hd['content'].apply(preprocess) 
dat_tt['content'] = dat_tt['content'].apply(preprocess) 
dat_bj['content'] = dat_bj['content'].apply(preprocess) 
dat_by['content'] = dat_by['content'].apply(preprocess) 
```

### K-means clustering
K-means initializes with a pre-determined number of clusters (We chose 3). Each observation is assigned to a cluster (cluster assignment) so as to minimize the within cluster sum of squares. Next, the mean of the clustered observations is calculated and used as the new cluster centroid. Then, observations are reassigned to clusters and centroids recalculated in an iterative process until the algorithm reaches convergence.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score

documents_hd = dat_hd['content'].values.tolist()
vectorizer = TfidfVectorizer(stop_words='english')
X_hd = vectorizer.fit_transform(documents_hd)

#This parameter decide the number of clusters you want
true_k = 3

model_hd = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model_hd.fit(X_hd)

print("Top terms per cluster:")
order_centroids_hd = model_hd.cluster_centers_.argsort()[:, ::-1]
terms_hd = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids_hd[i, :10]:
        print(' %s' % terms_hd[ind]),
    print

# get the cluster result of each tweet
X_hd = vectorizer.transform(documents_hd)
predictionX_hd = pd.DataFrame(model_hd.predict(X_hd),columns=['cluster'])
```

```python
documents_tt = dat_tt['content'].values.tolist()
vectorizer = TfidfVectorizer(stop_words='english')
X_tt = vectorizer.fit_transform(documents_tt)
model_tt = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model_tt.fit(X_tt)
# get the cluster result of each tweet
X_tt = vectorizer.transform(documents_tt)
predictionX_tt = pd.DataFrame(model_tt.predict(X_tt),columns=['cluster'])
```

```python
documents_bj = dat_bj['content'].values.tolist()
vectorizer = TfidfVectorizer(stop_words='english')
X_bj = vectorizer.fit_transform(documents_bj)
model_bj = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model_bj.fit(X_bj)

# get the cluster result of each tweet
X_bj = vectorizer.transform(documents_bj)
predictionX_bj = pd.DataFrame(model_bj.predict(X_bj),columns=['cluster'])
```

```python
documents_by = dat_by['content'].values.tolist()
vectorizer = TfidfVectorizer(stop_words='english')
X_by = vectorizer.fit_transform(documents_by)
model_by = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model_by.fit(X_by)

# get the cluster result of each tweet
X_by = vectorizer.transform(documents_by)
predictionX_by = pd.DataFrame(model_by.predict(X_by),columns=['cluster'])
```

```python
# merge the result into our dataset
cluster_dat_hd =  pd.concat([dat_hd,predictionX_hd], axis=1, sort=False)
cluster_dat_tt =  pd.concat([dat_hd,predictionX_tt], axis=1, sort=False)
cluster_dat_bj =  pd.concat([dat_hd,predictionX_bj], axis=1, sort=False)
cluster_dat_by =  pd.concat([dat_hd,predictionX_by], axis=1, sort=False)

cluster_dat_hd.head()
```
